```python
import pandas as pd
import os
```

```python
# Step 1: List all CSV files in the directory
directory = "trackingData"
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# Step 2: Initialize an empty DataFrame to store merged data
merged_df = pd.DataFrame()

# Step 3: Loop through each CSV file, merge with the previous DataFrame, and filter rows
for file in csv_files:
    df = pd.read_csv(os.path.join(directory, file))
    merged_df = pd.concat([merged_df, df])

# Step 4: Filter rows where frameId equals 1
merged_df_filtered = merged_df[merged_df['frameId'] == 1]
```

```python
merged_df_filtered.reset_index(drop=True, inplace=True)
```

```python
columns_to_drop = ['displayName','jerseyNumber','time','club','s','a','dis','o','dir','event']
merged_df_filtered = merged_df_filtered.drop(columns=columns_to_drop)
```

```python
merged_df_filtered.dropna(inplace=True)
```

```python
new_columns = []
for index, row in merged_df_filtered.iterrows():
    if row['playDirection'] == 'left':
        new_columns.append(row['x'])
    else:
        new_columns.append(120 - row['x'])
```

```python
merged_df_filtered['distanceToEndZone'] = new_columns
```

```python
merged_df_filtered.to_csv("cleanedTrackingData.csv", index=False)
```
