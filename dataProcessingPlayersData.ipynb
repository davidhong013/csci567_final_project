```python
import pandas as pd
import os
```

```python
playersData = pd.read_csv("players.csv")
playersRating = pd.read_csv('Madden 23 Player Ratings.csv')
```

```python
playersData
```

```python
playersRating
```

```python
# Assuming 'playerData' and 'playerRating' are your DataFrames
# Example: Merge 'playerData' and 'playerRating' on 'FullName' and 'displayName'

# Replace 'playerData' and 'playerRating' with the names of your DataFrames

# Merge 'playerData' and 'playerRating' on 'FullName' and 'displayName'
new_dataframe = playersData.merge(playersRating, left_on='displayName', right_on='Full Name', how='left')

# Display the new DataFrame
new_dataframe
```

```python
new_dataframe = new_dataframe.dropna(subset=['overall_rating'])
new_dataframe.columns
```

```python
columns_to_keep = ['height_x', 'weight_x','collegeName','Full Name', 'Team', 'Position','yearsPro','overall_rating','teamId','nflId',
                   'age',]
new_dataframe = new_dataframe.loc[:, columns_to_keep]
new_dataframe
```

```python
new_dataframe.to_csv("playersDataClean.csv")
```
